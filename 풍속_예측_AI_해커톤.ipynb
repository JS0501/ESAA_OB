### **3. 차별점 및 배울점**
#### **차별점**
상관계수를 살펴보고 변수 선택으로 다중공선성 문제를 해결하고자 했다.

Stacking에서 ExtraTreesRegressor를 사용했다.
#### **배울점**

**다중공선성**

- 독립 변수 간 상관계수가 높으면(다중공선성) 작은 데이터 변동에도 크게 변동한다.

- 표준 오차가 커져 t 검정 값이 작아지고 통계적으로 유의하지 않게 나올 가능성이 높아진다.

**ExtraTreesRegressor**

- 여러 개의 결정트리를 랜덤하게 만들어 평균을 내는 알고리즘이다.

- 랜덤 포레스트와 비슷하지만 트리 분할 방식이 더 무작위적이라는 차이가 있다.

- 계산 속도를 빠르게 하면서도 과적합을 방지하는 장점이 있다.
